```python
import pandas as pd
import seaborn as sns
```

## Trim and downsample

#### Constant current

```python
filename = "data/T05/0_05C/1.csv"
df = pd.read_csv(filename)

sns.scatterplot(x="time_s", y="voltage_v", data=df)
```

```python
sns.scatterplot(x="current_a", y="voltage_v", data=df)
```

```python
# find index of first negative current
first_neg_curr_idx = df.where(df["current_a"]<0).first_valid_index()
# find last index where voltage is 4.2 before first negative current
trim_idx = df.iloc[0:first_neg_curr_idx:1, :].where(df["voltage_v"]==4.2).last_valid_index()
# trim from fully charged state to end of test (i.e. 2.5V)
df = df.iloc[trim_idx:, :]
# reset time indices to start from 0 at fully charged state
df = df.reset_index(drop=True).reset_index().drop(columns="time_s").rename(columns={"index": "time_s"})

sns.scatterplot(x="time_s", y="voltage_v", data=df)
```
